Load in the "states_edu.csv" dataset and take a look at the head of the data

<h2> Feature Engineering </h2>

After exploring the data, you can choose to modify features that you would use to predict the performance of the students on your chosen response variable. 

You can also create your own features. For example, perhaps you figured that maybe a state's expenditure per student may affect their overall academic performance so you create a expenditure_per_student feature.

Use this space to modify or create features.

```python
df["Instruction_Expenditure_per_student"] = df["INSTRUCTION_EXPENDITURE"] / df["GRADES_ALL_G"]
df["Instruction_Expenditure_proportion"] = df["INSTRUCTION_EXPENDITURE"] / df["TOTAL_REVENUE"]
```

Feature engineering justification: It's a reasonable guess that both the amount of instruction expenditure spent on each student and proportion of investment on instruction indicate how important education is valued, and thus is assumed to have a positive correlation with students' academic performance.

<h2>Visualization</h2>

Investigate the relationship between your chosen response variable and at least two predictors using visualizations. Write down your observations.

**Visualization 1**

```python
df.plot.scatter(x='Instruction_Expenditure_per_student', y='AVG_MATH_8_SCORE', alpha=0.6)
plt.xlabel('Instruction Expenditure per student')
plt.ylabel('8th Grade Math Score')
```

**It seems like they indeed have a positive correlation.**

**Visualization 2**

```python
df.plot.scatter(x='Instruction_Expenditure_proportion', y='AVG_MATH_8_SCORE', alpha=0.6)
plt.xlabel('Proportion of investment on Instruction')
plt.ylabel('8th Grade Math Score')
```

**It seems like there is a weak positive correlation.**

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[["Instruction_Expenditure_per_student", "Instruction_Expenditure_proportion", "AVG_MATH_4_SCORE"]].dropna()
y = df.loc[X.index]["AVG_MATH_8_SCORE"]
```

```python
y.fillna(y.median(), inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=.3, random_state=42)
```

<h2> Prediction </h2>

ML Models [Resource](https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a)

```python
# import your sklearn class here
from sklearn.linear_model import LinearRegression
```

```python
# create your model here
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

## Evaluation

Choose some metrics to evaluate the performance of your model, some of them are mentioned in the tutorial.

```python
model.score(X_test, y_test)
```

```python
print(model.intercept_)
print(model.coef_)
```

We have copied over the graphs that visualize the model's performance on the training and testing set. 

Change `col_name` and modify the call to `plt.ylabel()` to isolate how a single predictor affects the model.

```python
col_name = 'Instruction_Expenditure_per_student'

f = plt.figure(figsize=(12,6))
plt.scatter(X_train[col_name], y_train, color = "red")
plt.scatter(X_train[col_name], model.predict(X_train), color = "green")

plt.legend(['True Training','Predicted Training'])
plt.xlabel(col_name)
plt.ylabel('AVG_MATH_8_SCORE')
plt.title("Model Behavior On Training Set")
```

```python
col_name = 'Instruction_Expenditure_per_student'

f = plt.figure(figsize=(12,6))
plt.scatter(X_test[col_name], y_test, color = "blue")
plt.scatter(X_test[col_name], model.predict(X_test), color = "black")

plt.legend(['True testing','Predicted testing'])
plt.xlabel(col_name)
plt.ylabel('AVG_MATH_8_SCORE')
plt.title("Model Behavior on Testing Set")
```

We want to explore the relationship between instruction expenditure and academic performance. For academic performance , we just choose the average math score of 8th grade students. The graph shows the former has a strong positive relation and the latter also show a positive relation but weaker. Also, we include the 4th grade average of math score as a predictor, because intuitively, students who do well when they are young are likely to keep exceling. Our model did a good job for its accuracy score is 0.845. From visualization of our model, we can also see that our model captures the pattern both for training set and test set, so it won't be considered as overfitting.
